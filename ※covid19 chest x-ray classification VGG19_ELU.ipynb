# 파일경로 불러오기

```python
import os
```

```python
#normal이라는 directory하위에 있는 모든 이미지들을 list형태로 반환한다.
IMAGE_FORMAT = ["jpeg", 'jpg', 'png']

def list_image_files(data_dir, sub_dir):
    image_files = []
    images_dir = os.path.join(data_dir, sub_dir)
    for file_path in os.listdir(images_dir):
        if file_path.split(".")[-1] in IMAGE_FORMAT:
            image_files.append(os.path.join(sub_dir, file_path))
    return image_files
```

# 이미지파일을 RGB 3차원 배열로 불러오기

```python
import cv2
```

```python
#데이터 directory의 file path와 연결을 시켜주고, opencv에서 이미지형태로 불러올수 있도록 파일의 경로를 불러온다.
#BGR형태를 RGB형태로 바꿔준다.
def get_RGB_image(data_dir,file_name):
    image_file = os.path.join(data_dir, file_name)
    image = cv2.imread(image_file)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    return image
```

# 학습데이터셋 클래스 구축

dataset instance를 생성될때
즉, data를 load하는 dataset은 기본적인 attribute를 가진다.
1. 생성자

data_dir, transformer은 인자값, 총 변수는 5개다 self는 class 내에서는 공통적으로 호출될수 있는 변수이고, self가 아닌것은 class 내부에서만 사용된다.

transformer은 none으로 한것은 향후에 이미지를 tensor로 변환하고 augmentation기법을 적용할때 transformer로 instance를 전달한다.

2. 데이터셋 길이
3. 데이터셋에 호출하는 인덱스에 맞는 아이템. 즉, return값은 이미지와 타겟값

```python
import torch
```

```python
train_data_dir = 'D:/covid/train'
class_list= ['normal', 'covid']
```

```python
class Chest_dataset():
    def __init__(self, data_dir, transformer=None): 
        self.data_dir = data_dir
        normals = list_image_files(data_dir, 'normal') #하위폴더에 있는 이미지 파일을 불러들어오는 함수
        covids = list_image_files(data_dir, 'covid')
        
        self.files_path = normals + covids #각 파일의 list를 모두 더한것
        self.transformer = transformer
    
    def __len__(self):
        return len(self.files_path)
        
    def __getitem__(self, index): 
        image_file = os.path.join(self.data_dir, self.files_path[index])
        image =cv2.imread(image_file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        #index를 호출하면 그에맞는 image와 label을 반환하는 함수다. 해당인덱스 해당하는 파일의 경로를 설정을 하고 open cv의 함수로 이미지의 파일을 불러온다 그다음 BGR을 RGB로 변환한다.
        #타겟값은 class list의 해당하는 index값으로 받게된다. calss name을 정하고이미지 파일을 불러오고온다. 
        #폴더 하위의 이미지 데이터가 있었으니까 이미지 파일을 os.sep으로 분리를 해주고 첫번째 값을 class_name으로 설정한것이다.
        #calss name을 class list에서 대응하는 class index로 받아주게되면 class_id로 지정을 할 수있다.
        #normal인경우에는 class id가 0이 된다.
        
        class_name = self.files_path[index].split(os.sep)[0]
        class_id = class_list.index(class_name)
        
        #추후에 transformer을 하면 tensor형으로 변형이 되기때문에 변환을 해준다.
        #class id는 tensor로 바꾸고 long(int)값으로 변환해서 반환한다.
        if self.transformer:
            image = self.transformer(image)
            class_id = torch.Tensor([class_id]).long()
        
        return image, class_id
        #print(class_name, class_id)
        #print(class_name)
        
        #image filep path경로 확인
        #print(self.data_dir, self.files_path[index])
        #print(self.files_path[index].split(os.sep)[0])
        
        #target = class_list.index
        
```

```python
#두번쨰 방법 : #dict형태의 dataloaders라는 변수에 train과 val라는 key를 주어준다. 이것으로 dataloader를 한번에 반환한다.
def build_dataloader(train_data_dir, val_data_dir):
    dataloaders = {} 
    train_dset = Chest_dataset(train_data_dir, transformer)
    dataloaders['train'] = DataLoader(train_dset, batch_size=32, shuffle=True, drop_last=True) 
    val_dset = Chest_dataset(val_data_dir, transformer)
    dataloaders['val'] = DataLoader(val_dset, batch_size=32, shuffle=False)
    
    return dataloaders
```

```python
def build_vgg19_based_model(device='cuda'):
    #model = models.vgg19(pretrained=True)
    model.features[1] = nn.ELU()
    model.features[3] = nn.ELU()
    model.features[6] = nn.ELU()
    model.features[8] = nn.ELU()
    model.features[11] = nn.ELU()
    model.features[13] = nn.ELU()
    model.features[15] = nn.ELU()
    model.features[17] = nn.ELU()
    model.features[20] = nn.ELU()

    model.features[22] = nn.ELU()
    model.features[24] = nn.ELU()
    model.features[26] = nn.ELU()
    model.features[29] = nn.ELU()
    model.features[31] = nn.ELU()
    model.features[33] = nn.ELU()
    model.features[35] = nn.ELU()
    model.avgpool = nn.AdaptiveAvgPool2d(output_size=(1,1))
    model.classifier = nn.Sequential(
        nn.Flatten(),
        nn.Linear(512, 256),
        nn.ReLU(),
        #nn.Dropout(0.1), #네트워크 가중치 편향을 막기 위해 사용
        nn.Linear(256,2),
        nn.Sigmoid() #binary class를 예측할땐 sigmoid, 다중 분류 예측에는 softmax
    )
    return model.to(device) #작성된 모델을 cuda로 지정된 torch device에 mount하라는 뜻이다. 이것은 gpu 연산을 가능하게한다.
```

# 손실함수 (Cross Entropy)

classification의 경우으넨 0과 1사이의 값을 가지게 된다. binary crossentropy의 겨웅에도 베르누이 확률 분포를 따르기 때문에 y는 0과 1이 들어가고, y_pred는 모델의 결과값이 된다. 
Loss = (Y)(-log(Y_pred)) + (1-Y)(-log(1-Y_pred))
y가 target값이 1일때는 전체수식에서 앞에 항만 남고, 뒤에 항은 1-y가 0이 되기 때문에 제거된다. 반대로 y가 0의값을 가지게 될때 앞의 값은 제거되고, 뒤에 수식만 남게된다. 이때 y_pred값은 어떻게 이동해야할까?

# SGDM (Stochastic Gradient Descent with Momentum) Optimization

y축은 error, x축으s weight space를 가진다고할때, 모델의 파라미터는 error(loss fuction)을 최소화 하는방향으로 기울기를 하강시키기 위해 weight space를 탐색한다. 이론적으로는 local minima가 아닌 global minia를 찾는것이 opitmization이다.
local minima가 global minima인지 아닌지 딜레마에 빠질수 있는데, Momentum을 통해서 다른 더 낮은 local minima가 있는지 탐색하는 방법이 SGDM이다. momentum의 default 값은0이다.

** GD(Gradient Descent)와 SGD(Stochastic Gradient Descent)의 차이는 무엇일까?

- GD(Gradient Descent)는 모든 데이터에 대해서 학습을 하려고 하기때문에 모든 방향성을 보두 고려해서 error를 최소화하는 weight space를 탐색한다. 이럴 경우에는 근처에 있는 극소점에 도달하기 때문에 local minima에 빠지기 쉽다.

- SGD(Stochastic Gradient Descent)의 경우는 모든 데이터가 아닌 일부 데이터를 뽑아서 데이터를 학습하기 때문에 모든 방향이 아닌 일부 방향에 대해서 학습하게된다. 이때 일부 데이터의 개수는 batch_size가 된다. 이에따라 학습시간은 줄어들고 local minima에 빠질 확률이 줄어든다. 이때, momentum을 이용하면 다른 극소점까지 탐색하기 때문에 global minima에 도달할 확률이 커지게 된다.


```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)  #lr = 1e-3
```

# 모델 학습을 위한 함수 구현하기

```python
device = torch.device('cuda')
```

```python
# 한번 iteration할때 train과 validation이 동작을 한다. 한번 epoch할떄 train한번, epoch한번 돌아서 loss를 반환을 하고 그 loss를 토대로 overfitting이 되는지 한단한다.
def train_one_epoch(dataloaders, model, optimizer, loss_func, device):
    loss ={}
    
    for phase in["train", "val"]: 
        running_loss = 0.0 
        #중간 중간 loss를 확인하기 위해서 필요하다. 중간중간에 running loss를 받아와서 print문으로 출력해서 잘 학습이 되는지 확인할 수 있게한다.
        
        if phase == "train":
            model.train()
        else:
            model.eval() 
            #모델내 btach normalizization나 dropout등을 activation 할지말지 구별한다. 즉, train과 validation set에서 다르게 작동하는 layer가 있기 때문에 이를 구별하기위해 사용한다.
            
        for index, batch in enumerate(dataloaders[phase]): #dataloaders를 phase에 맞게 불러오고 index값과 batch형태로 iterative하게 가져온다. batch는 엄밀히 말하면 mini_batch다.
            image = batch[0].to(device) #image 
            label = batch[1].to(device) #label 위에선 class id normal이면 0 covid면1
            
            
            """
            with torch.no_grad(): 는 valid일때, set_grad_enabled는 tarin일떄 사용한다.
            다음 코드는 밑에 with torch.set_grad_enabled(phase =="train"):와 같다.
            if phase == 'train':
                prediction = model(image)
            elif phase == 'val':
                woth torch.no_grad():
                    prediction = model(image)
            """
            
            with torch.set_grad_enabled(phase =="train"):
                prediction = model(image)
                loss = loss_func(prediction, target)
                
                if phase == "train":
                    optimizer.zero_grad() #loss를 계산하고, loss에 대해서 optimizer의 weight값을 업데이트 할때, optimizer에 zero_grad를 안해주면 미분의 누적값들이 계속 생기게 된다. 따라서 학습이 원하는 방향으로 진행이 되지 않을수 있다. 그렇기 때문에 loss를 기반으로 optimizer를 수행하기전에 미분함수의 파라미터에 들어있는 모든 gradient를 0으로 만들어주고 그다음 loss에 해당하는 파라미터의 기울기 값을 다시 갱신을 해주고 다시 optimizer를 수행한다.
                    loss.backward() #loss가 구해지면 model의 파라미터의 gradient를 추적할수 있는 기능이 있는데 loss.backward()를 통해 loss값에 관련이 있는 모든 model의 파라미터의 미분값이 업데이트 되게 된다.
                    optimizer.step() #이 미분값을 optimizer에 전달하게되고, 가중치를 미리 설정한 수식에 따라서 가중치와 learning rate를 토대로 모델의 weight값을 갱신하게 된다. loss를 가지고 optimizer을 수행한다.
                    
            running_loss += loss.item() #loss 값의 value만 받아서 아까 설정한 running_loss에 누적을 시켜준다.
                            
                #index 10번 주기의 간격으로 running loss를 출력을 해라라는 의미다.
            if phase == "train":
                if index % 10 == 0:
                    print(f"{index}/{len(dataloaders['train'])} - Running loss: {loss.item()}")#running_loss의 값은 최종적으로 dataloader의 개수(batch크기의 평균한 값)을 계산하고 각 phase의 key값에 전달해준다.
                    
        losses[phase] = running_loss / len(dataloader[phase])
        
    return losses
```

- Top-5 accuracy

```python
@torch.no_grad()
def get_accuracy(image, target, model):
    batch_size = image.shape[0] #이미지에서 batch크기를 계산을 하고 이미지를 연산을 해서 output값에 넣어준다
    prediction = model(image) # prediction값은 (batch_size, num_classes)이다. 
    #예를들어 이때 각 class마다 [B, 0]: 0,1, [B,1]:0.2, [B,2]:0.7
    _, pred_label = torch.max(prediction, dim=1) # dimension 1을 기준으로 max값을 구한다.위의 경우에서 (B,1) -> pred_label =2
    is_correct = (pred_label == target) #pred_label과 target을 매칭하여 True/ False를 반환
    accuracy = is_correct.cpu().numpy().sum() / batch_size #True, False를 모두 더하면 false는0 true는 1로 대변이 된다. 이것을 batch_size로 나눈다.
    return accuracy

#accuracy를 함수로 작성한 이유는 매 iteration, epoch에 train과 validation에 대해서 매번 accuracy를 연산을 하고, 개선되는 과정을 살펴 볼것이기 때문에 이 함수가 반복적으로 호출되기 위해 함수를짯다.
#accuracy가 계산되는 과정에서는 backward과정이 필요없기 때문에 decorater까지 넣어주어 model의 파라미터에 대해서 gradient update 과정을 꺼준다.
```

```python
# 한번 iteration할때 train과 validation이 동작을 한다. 한번 epoch할떄 train한번, epoch한번 돌아서 loss를 반환을 하고 그 loss를 토대로 overfitting이 되는지 한단한다.
def train_one_epoch(dataloader, model, optimizer, loss_function, device):
    losses = {}
    accuracies = {}
    
    for phase in["train", "val"]: 
        running_loss = 0.0
        running_correct = 0.0
        #중간 중간 loss를 확인하기 위해서 필요하다. 중간중간에 running loss를 받아와서 print문으로 출력해서 잘 학습이 되는지 확인할 수 있게한다.
        
        if phase == "train":
            model.train()
        else:
            model.eval() 
            #모델내 btach normalizization나 dropout등을 activation 할지말지 구별한다. 즉, train과 validation set에서 다르게 작동하는 layer가 있기 때문에 이를 구별하기위해 사용한다.
            
        for index, batch in enumerate(dataloader[phase]): #dataloaders를 phase에 맞게 불러오고 index값과 batch형태로 iterative하게 가져온다. batch는 엄밀히 말하면 mini_batch다.
            image = batch[0].to(device) #image 
            label = batch[1].squeeze(dim=1).to(device) #label 위에선 class id normal이면 0 covid면1
            
            with torch.set_grad_enabled(phase =="train"):
                prediction = model(image)
                loss = loss_func(prediction, label)
                
                optimizer.zero_grad()
                
                if phase == "train":
                    #optimizer.zero_grad() #loss를 계산하고, loss에 대해서 optimizer의 weight값을 업데이트 할때, optimizer에 zero_grad를 안해주면 미분의 누적값들이 계속 생기게 된다. 따라서 학습이 원하는 방향으로 진행이 되지 않을수 있다. 그렇기 때문에 loss를 기반으로 optimizer를 수행하기전에 미분함수의 파라미터에 들어있는 모든 gradient를 0으로 만들어주고 그다음 loss에 해당하는 파라미터의 기울기 값을 다시 갱신을 해주고 다시 optimizer를 수행한다.
                    loss.backward() #loss가 구해지면 model의 파라미터의 gradient를 추적할수 있는 기능이 있는데 loss.backward()를 통해 loss값에 관련이 있는 모든 model의 파라미터의 미분값이 업데이트 되게 된다.
                    optimizer.step() #이 미분값을 optimizer에 전달하게되고, 가중치를 미리 설정한 수식에 따라서 가중치와 learning rate를 토대로 모델의 weight값을 갱신하게 된다. loss를 가지고 optimizer을 수행한다.
                    
            running_loss += loss.item() #loss 값의 value만 받아서 아까 설정한 running_loss에 누적을 시켜준다.
            running_correct += get_accuracy(image, label, model)
                            
                #index 10번 주기의 간격으로 running loss를 출력을 해라라는 의미다.
            if phase == "train":
                if index % 10 == 0:
                    print(f"{index}/{len(dataloaders['train'])} - Running loss: {loss.item()}")#running_loss의 값은 최종적으로 dataloader의 개수(batch크기의 평균한 값)을 계산하고 각 phase의 key값에 전달해준다.
                    
        losses[phase] = running_loss / len(dataloader[phase])
        accuracies[phase] = running_correct / len(dataloader[phase])
    return losses, accuracies
```

```python
# 한번 iteration할때 train과 validation이 동작을 한다. 한번 epoch할떄 train한번, epoch한번 돌아서 loss를 반환을 하고 그 loss를 토대로 overfitting이 되는지 한단한다.
def train_one_epoch(dataloader, model, optimizer, loss_function, device):
    losses = {}
    accuracies = {}
    
    for phase in["train", "val"]: 
        running_loss = 0.0
        running_correct = 0.0
        #중간 중간 loss를 확인하기 위해서 필요하다. 중간중간에 running loss를 받아와서 print문으로 출력해서 잘 학습이 되는지 확인할 수 있게한다.
        
        if phase == "train":
            model.train()
        else:
            model.eval() 
            #모델내 btach normalizization나 dropout등을 activation 할지말지 구별한다. 즉, train과 validation set에서 다르게 작동하는 layer가 있기 때문에 이를 구별하기위해 사용한다.
            
        for index, batch in enumerate(dataloader[phase]): #dataloaders를 phase에 맞게 불러오고 index값과 batch형태로 iterative하게 가져온다. batch는 엄밀히 말하면 mini_batch다.
            image = batch[0].to(device) #image 
            label = batch[1].squeeze(dim=1).to(device) #label 위에선 class id normal이면 0 covid면1
            
            with torch.set_grad_enabled(phase =="train"):
                prediction = model(image)
                loss = loss_func(prediction, label)
                
                optimizer.zero_grad()
                
                if phase == "train":
                    #optimizer.zero_grad() #loss를 계산하고, loss에 대해서 optimizer의 weight값을 업데이트 할때, optimizer에 zero_grad를 안해주면 미분의 누적값들이 계속 생기게 된다. 따라서 학습이 원하는 방향으로 진행이 되지 않을수 있다. 그렇기 때문에 loss를 기반으로 optimizer를 수행하기전에 미분함수의 파라미터에 들어있는 모든 gradient를 0으로 만들어주고 그다음 loss에 해당하는 파라미터의 기울기 값을 다시 갱신을 해주고 다시 optimizer를 수행한다.
                    loss.backward() #loss가 구해지면 model의 파라미터의 gradient를 추적할수 있는 기능이 있는데 loss.backward()를 통해 loss값에 관련이 있는 모든 model의 파라미터의 미분값이 업데이트 되게 된다.
                    optimizer.step() #이 미분값을 optimizer에 전달하게되고, 가중치를 미리 설정한 수식에 따라서 가중치와 learning rate를 토대로 모델의 weight값을 갱신하게 된다. loss를 가지고 optimizer을 수행한다.
                    
            running_loss += loss.item() #loss 값의 value만 받아서 아까 설정한 running_loss에 누적을 시켜준다.
            running_correct += get_accuracy(image, label, model)
                            
                #index 10번 주기의 간격으로 running loss를 출력을 해라라는 의미다.
            if phase == "train":
                if index % 10 == 0:
                    print(f"{index}/{len(dataloaders['train'])} - Running loss: {loss.item()}")#running_loss의 값은 최종적으로 dataloader의 개수(batch크기의 평균한 값)을 계산하고 각 phase의 key값에 전달해준다.
                    
        losses[phase] = running_loss / len(dataloader[phase])
        accuracies[phase] = running_correct / len(dataloader[phase])
    return losses, accuracies
```

```python
num_epochs = 10

best_acc = 0.0
train_loss, train_acc = [], []
val_loss, val_acc = [], []

for epoch in range(num_epochs):
    losses, accuracies = train_one_epoch(dataloaders, model, optimizer, loss_func, device)
    train_loss.append(losses['train'])
    train_acc.append(accuracies['train'])
    val_loss.append(losses['val'])
    val_acc.append(accuracies['val'])
    
    print(f"{epoch+1}/{num_epochs}-Train loss:{losses['train']}, Val loss:{losses['val']}")
    print(f"{epoch+1}/{num_epochs}-Train acc:{accuracies['train']}, Val acc:{accuracies['val']}")
    
    if accuracies['val'] > best_acc:
        best_acc = accuracies['val']
        torch.save(model.state_dict(), f"vgg_model_{epoch}.pth")
```

```python
# DenseNet BottleNeck
class BottleNeck(nn.Module):
    def __init__(self, in_channels, growth_rate):
        super().__init__()
        inner_channels = 4 * growth_rate

        self.residual = nn.Sequential(
            nn.BatchNorm2d(in_channels),
            nn.ReLU(),
            nn.Conv2d(in_channels, inner_channels, 1, stride=1, padding=0, bias=False),
            nn.BatchNorm2d(inner_channels),
            nn.ReLU(),
            nn.Conv2d(inner_channels, growth_rate, 3, stride=1, padding=1, bias=False)
        )

        self.shortcut = nn.Sequential()

    def forward(self, x):
        return torch.cat([self.shortcut(x), self.residual(x)], 1)
```

```python
# Transition Block: reduce feature map size and number of channels
class Transition(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()

        self.down_sample = nn.Sequential(
            nn.BatchNorm2d(in_channels),
            nn.ReLU(),
            nn.Conv2d(in_channels, out_channels, 1, stride=1, padding=0, bias=False),
            nn.AvgPool2d(2, stride=2)
        )

    def forward(self, x):
        return self.down_sample(x)
```

```python
# DenseNet
class DenseNet(nn.Module):
    def __init__(self, nblocks, growth_rate=12, reduction=0.5, num_classes=2, init_weights=True):
        super().__init__()

        self.growth_rate = growth_rate
        inner_channels = 2 * growth_rate # output channels of conv1 before entering Dense Block

        self.conv1 = nn.Sequential(
            nn.Conv2d(3, inner_channels, 7, stride=2, padding=3),
            nn.MaxPool2d(3, 2, padding=1)
        )

        self.features = nn.Sequential()

        for i in range(len(nblocks)-1):
            self.features.add_module('dense_block_{}'.format(i), self._make_dense_block(nblocks[i], inner_channels))
            inner_channels += growth_rate * nblocks[i]
            out_channels = int(reduction * inner_channels)
            self.features.add_module('transition_layer_{}'.format(i), Transition(inner_channels, out_channels))
            inner_channels = out_channels 
        
        self.features.add_module('dense_block_{}'.format(len(nblocks)-1), self._make_dense_block(nblocks[len(nblocks)-1], inner_channels))
        inner_channels += growth_rate * nblocks[len(nblocks)-1]
        self.features.add_module('bn', nn.BatchNorm2d(inner_channels))
        self.features.add_module('relu', nn.ReLU())

        self.avg_pool = nn.AdaptiveAvgPool2d((1,1))
        self.linear = nn.Linear(inner_channels, num_classes)

        # weight initialization
        if init_weights:
            self._initialize_weights()
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.features(x)
        x = self.avg_pool(x)
        x = x.view(x.size(0), -1)
        x = self.linear(x)
        return x

    def _make_dense_block(self, nblock, inner_channels):
        dense_block = nn.Sequential()
        for i in range(nblock):
            dense_block.add_module('bottle_neck_layer_{}'.format(i), BottleNeck(inner_channels, self.growth_rate))
            inner_channels += self.growth_rate
        return dense_block

    def _initialize_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.constant_(m.weight, 1)
                nn.init.constant_(m.bias, 0)
            elif isinstance(m, nn.Linear):
                nn.init.normal_(m.weight, 0, 0.01)
                nn.init.constant_(m.bias, 0)

def DenseNet_121():
    return DenseNet([6, 12, 24, 6])
```
